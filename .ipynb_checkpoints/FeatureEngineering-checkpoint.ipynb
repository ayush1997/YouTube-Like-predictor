# Feature engineering/derivation
### Some new feature are derived from the present attributes.

## Available features

```python
df_new.columns
```

### 1. Video Description Length

```python
df_new["description_length"] = df_new.description.apply(lambda x : len(x.split(" ")))
```

### 2. Links to other website(http) in description

```python
def find_http(lst):
    return len([1 for i in lst if 'http' in i])
    
df_new["http_in_descp"] = df_new.description.apply(lambda x : find_http(x.split(" ")))
```

### 3. No of Tags in description 

```python
def tag_in_desc(df):
    
    if df.tags != None :
        a = [x.lower() for x in df.tags]
        b = df.description.lower()
        return len([i for i in a if i in b])
    else:
        return 0
df_new["tags_in_desc"] = df_new.apply(tag_in_desc,axis=1)
```

### 4. Video title length

```python
df_new["video_title_length"] = df_new.title.apply(lambda x : len(x.split(" ")))
```

### 5. No. of Tags in title

```python
def tag_in_title(df):
    if df.tags != None :
        a = [x.lower() for x in df.tags]
        b = df.title.lower()
        return len([i for i in a if i in b])
    else:
        return 0
df_new["tags_in_title"] = df_new.apply(tag_in_desc,axis=1)
```

### 6. No of tags assigned to video

```python
def len_tags(x):
    if x ==None:
        return 0
    else:
        return len(x)
df_new["No_of_tags"] = df_new.tags.apply(lambda x : len_tags(x))
```

### 7. Channel title length

```python
df_new["channel_title_length"] = df_new.channelTitle.apply(lambda x : len(x.split(" ")))
```

### 8. Channel description length

```python
df_new["channel_description_length"] = df_new.ChannelDescription.apply(lambda x : len(x.split(" ")))
```

### 9. How old is video (in months) 

```python
def get_months(x):
    return 12-x.month + 1 + (2016 - (x.year+1)+1)*12
df_new["months_old"] = pd.to_datetime(df_new.publishedAt).apply(lambda x: get_months(x))
```

### 10. How old in channel

```python
def get_months(x):
    return 12-x.month + 1 + (2016 - (x.year+1)+1)*12
df_new["channel_months_old"] = pd.to_datetime(df_new.ChannelPublishedAt).apply(lambda x: get_months(x))
```

### 11. Get day of video upload

```python
import datetime
def get_day(x):
    return datetime.date(x.year,x.month,x.day).strftime("%A")
df_new["day_uploaded"] = pd.to_datetime(df_new.publishedAt).apply(lambda x: get_day(x))
```

### The columns which are of no use further are dropped

```python
df_new = df_new.drop([u'index',u'level_0',u'description', u'tags', u'title',u'publishedAt',u'ChannelDescription', u'ChannelPublishedAt', u'channelTitle'],axis=1)
```

### ViewCount/channel month old

```python
df_new["viewCount/channel_month_old"] = df_new.apply(lambda x: (x["viewCount"])/x["channel_months_old"],axis=1)
```

### ViewCount/ Video month old

```python
df_new["viewCount/video_month_old"] = df_new.apply(lambda x: (x["viewCount"])/x["months_old"],axis=1)
```

### ViewCount / http_in_descp

```python
df_new["viewCount/http_in_descp"] = df_new.apply(lambda x: (x["viewCount"]+1)/(x["http_in_descp"]+1),axis=1)
```

### ViewCount / No of tags

```python
df_new["viewCount/NoOfTags"] = df_new.apply(lambda x: (x["viewCount"]+1)/(x["No_of_tags"]+1),axis=1)
```

### subscriberCount / videoCount

```python
df_new["subscriberCount/videoCount"] = df_new.apply(lambda x: (x["channel_subscriberCount"]+1)/(x["channel_videoCount"]+1),axis=1)
```

### ViewCount/ ( tags_in_desc  + tags_in_title)

```python
df_new["viewCount/tags_in_desc"] = df_new.apply(lambda x:( x["viewCount"])/(x["tags_in_desc"] + x["tags_in_desc"]+1),axis=1)
```

### ChannelViewCount / channelVideoCount

```python
df_new["channelViewCount/channeVideoCount"] = df_new.apply(lambda x:( x["channel_ViewCount"]+1)/(x["channel_videoCount"]+1),axis=1)
```

### ChannelViewCount / socialLink

```python
df_new["channelViewCount/socialLink"] = df_new.apply(lambda x:( x["channel_ViewCount"]+1)/(x["social_links"]+1),axis=1)
```
